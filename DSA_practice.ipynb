```python
# Q: Best time to buy and sell stocks

price = [7,1,5,3,6,4]

def max_profit(price):
    max_diff = 0
    buy = 0
    sell = 0

    for i in range(len(price)):
        for j in range(len(price)):
            if i == j:
                pass 
            else:
                if i < j:
                    diff = price[j] - price[i]
                    if diff > max_diff:
                        max_diff = diff
                        buy = price[i]
                        sell = price[j]
                
    print(max_diff,buy, sell)
```

```python
max_profit(price)
```
